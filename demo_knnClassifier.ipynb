```python
from classification_algorithms.knn_classifier import KNNClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### generate data and plot data

```python
def generate_random_points(size=10, low=0, high=1):
    ''' Generate a set of random 2D points
        size -- number of points to generate
        low  -- min value
        high -- max value
    '''
    return (high - low) * np.random.random_sample((size, 2)) + low
```

```python
X1 = generate_random_points(20, 0, 1)
X2 = generate_random_points(20, 1, 2)
y = np.hstack((np.zeros(20), np.ones(20)))
X = np.vstack((X1, X2))
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], cmap='YlOrRd', c=y, s=10)
plt.show()
```

### use knn classifier

```python
knn = KNNClassifier()
knn.fit(X, y)
```

```python
x1 = np.linspace(X[:,0].min(), X[:,0].max(), 100)
x2 = np.linspace(X[:,1].min(), X[:,1].max(), 100)
x1, x2 = np.meshgrid(x1, x2)
y_hat = knn.predict(np.c_[x1.flatten(), x2.flatten()], k=5).reshape(x1.shape)
```

```python
plt.figure(figsize=(7, 5))
plt.contourf(x1, x2, y_hat)
plt.scatter(X[:, 0], X[:, 1], cmap='YlOrRd', c=y, s=10)
plt.xlim([X[:,0].min(), X[:,0].max()])
plt.ylim([X[:,1].min(), X[:,1].max()])
plt.show()
```
